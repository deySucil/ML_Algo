```python
import pandas as pd
import yfinance as yf
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing_Price.csv')
df = df.iloc[0:10,0:4]
df
# print(df.columns)
```

```python
# %matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.plot(df.area, df.price, c='red')
plt.show()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(new_df, df.price)
```

```python
predict_line = model.predict(new_df)
plt.scatter(df.area, df.price, c='red')
plt.plot(df.area, predict_line, c='blue')
```

```python
model.predict([[700]])
```

```python
model.coef_
```

```python
model.intercept_
```
