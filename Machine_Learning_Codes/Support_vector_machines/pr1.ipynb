```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
len(iris.data)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target'] = iris.target
df[df.target==1].head()
```

```python
df['flower_name'] = df.target.apply(lambda x:iris.target_names[x])
df[df.target==2].head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Sepal_length')
plt.ylabel('Sepal_width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],c='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],c='red', marker='.')
```

```python
plt.xlabel('Petal_length')
plt.ylabel('Patel_width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],c='green',marker='+')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],c='red',marker='.')
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['target','flower_name'],axis='columns')
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model_c = SVC(C=10)
model_c.fit(x_train,y_train)
model_c.score(x_test,y_test)
```

```python
model_g = SVC(gamma=5)
model_g.fit(x_train,y_train)
model_g.score(x_test,y_test)
```

```python
model_linearKernel = SVC(C=10,kernel='linear')
model_linearKernel.fit(x_train,y_train)
model_linearKernel.score(x_test,y_test)
```
