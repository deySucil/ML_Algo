```python
import pandas as pd
import yfinance as yf
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df.year,df['per capita income (US$)'],c='red',marker='+')
```

```python
model = linear_model.LinearRegression()

X = df.drop('per capita income (US$)', axis='columns')
model.fit(X,df['per capita income (US$)'])
```

```python
Y = model.predict(X)
plt.scatter(df.year, df['per capita income (US$)'], c='red', marker='+')
plt.plot(df.year,Y,c='blue')
```
